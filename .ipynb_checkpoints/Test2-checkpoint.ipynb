### Transform Eth DataFrame

```python
# Create a filtered dataframe from specific columns
Eth_cols = ["date", "open", "close"]
Eth_transformed= Eth_df[Eth_cols].copy()

Eth_transformed.head()
```

```python
DOW_cols = ["Date", "Close"]
DOW_transformed2= DOW_df[DOW_cols].copy()

DOW_transformed2.head()
```

```python
Eth_cols = ["date", "close"]
Eth_transformed2= Eth_df[Eth_cols].copy()

Eth_transformed2.head()
```

```python
Eth_transformed.reset_index()
```

```python
for ind in Eth_transformed2['date'].index:
    n = Eth_transformed2['date'][ind]
    tmp = str(n)
    

    dt = datetime.datetime.strptime(tmp,'%Y-%m-%d %H:%M:%S')
    
    dt.date()

    dd = dt.date()
    
    Eth_transformed2['date'][ind] = dd
    
    print(dd)
    print("-")
```

```python
Eth_transformed2.head()
```

```python
Eth_transformed2[Eth_transformed2.date != '2018-12-31']
```

```python
eth_cols = ["date", "close"]
eth_transformed= Eth_transformed2[eth_cols].copy()

# Rename the column headers
eth_transformed = eth_transformed.rename(columns={"date": "date", "close": "Eth_Close"})

# Clean the data by dropping duplicates and setting the index
eth_transformed.drop_duplicates("Eth_Close", inplace=True)
eth_transformed.set_index("Eth_Close", inplace=True)

eth_transformed.head()
```

```python
dow_cols = ["Date", "Close"]
dow_transformed= DOW_transformed2[dow_cols].copy()

# Rename the column headers
dow_transformed = dow_transformed.rename(columns={"Date": "Date", "Close": "DOW_Close"})

# Clean the data by dropping duplicates and setting the index
dow_transformed.drop_duplicates("DOW_Close", inplace=True)
dow_transformed.set_index("DOW_Close", inplace=True)

dow_transformed.head()
```
